```python
from google.colab import drive
import os
import time
import shutil

drive.mount('/content/drive')
SAVE_DIR = "/content/drive/MyDrive/PairLIE_checkpoints/"  # 원하는 저장 경로 설정

if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

WEIGHT_DIR = "/content/PairLIE/weights/"  # main.py에서 모델을 저장하는 로컬 경로

# 체크포인트 백업
if os.path.exists(WEIGHT_DIR):
    files = sorted(os.listdir(WEIGHT_DIR))  # 최신 파일 찾기
    for file in files:
        local_path = os.path.join(WEIGHT_DIR, file)
        drive_path = os.path.join(SAVE_DIR, file)

        if not os.path.exists(drive_path):  # 중복 저장 방지
            shutil.copy(local_path, drive_path)
            print(f"✅ Checkpoint saved to Google Drive: {drive_path}")
```

```python
# 1️⃣ 최신 체크포인트 찾기
def load_checkpoint():
    checkpoint_files = sorted(os.listdir(SAVE_DIR))
    if not checkpoint_files:
        print("⚠ No checkpoint found, starting from scratch.")
        return None

    latest_checkpoint = checkpoint_files[-1]  # 가장 최신 체크포인트 선택
    checkpoint_path = os.path.join(SAVE_DIR, latest_checkpoint)
    print(f"✅ Loading checkpoint: {checkpoint_path}")

    checkpoint = torch.load(checkpoint_path)
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])

    start_epoch = checkpoint['epoch'] + 1
    return start_epoch

# 2️⃣ 체크포인트 불러오기
start_epoch = load_checkpoint()
if start_epoch:
    print(f"✅ Resuming training from epoch {start_epoch}")
else:
    start_epoch = 1  # 체크포인트가 없으면 1부터 시작
```

```python
import torch
from PIL import Image
import numpy as np
from torchvision import transforms
import matplotlib.pyplot as plt
from google.colab import files
from net.net import net

# 1. 이미지 업로드
uploaded = files.upload()

# 업로드된 파일 경로 얻기
image_path = list(uploaded.keys())[0]
image = Image.open(image_path)

# 2. 이미지를 RGB로 변환하고 numpy 배열로 변환
image = image.convert('RGB')  # 이미지를 RGB로 변환 (이미지가 흑백일 경우, 컬러로 변환)
image_np = np.array(image)

# 이미지 차원 확인 (예: [H, W, C])
print(f"Original Image Shape: {image_np.shape}")

# 3. 이미지를 tensor로 변환 (전처리)
transform = transforms.Compose([
    transforms.ToTensor(),  # 텐서로 변환
])
image_tensor = transform(image).unsqueeze(0)  # 배치 차원 추가 (1, 3, H, W)

# 4. 모델 정의 및 불러오기
model = net().cuda()  # 'net'은 모델을 정의한 클래스입니다. 이를 불러와야 합니다.
model_path = 'weights/epoch_400.pth'  # 모델 경로
model.load_state_dict(torch.load(model_path))
model.eval()  # 모델을 평가 모드로 설정

# 5. 모델에 입력하여 결과 얻기
image_tensor = image_tensor.cuda()  # GPU 사용 시 텐서 GPU로 이동
output = model(image_tensor)  # 모델에 입력하여 출력 얻기

# 모델이 반환하는 값이 튜플이라면, 첫 번째 값을 사용 (출력 값은 모델에 따라 다를 수 있음)
output_image = output[0]  # 모델의 첫 번째 출력 값 가져오기

# 6. 출력 이미지 처리
output_image = output_image.squeeze(0)  # 배치 차원 제거 (1, C, H, W) -> (C, H, W)
output_image = output_image.squeeze(0)  # 채널 차원 제거 (C, H, W) -> (H, W)

# 7. 출력 이미지 차원 확인
print(f"Output Image Shape: {output_image.shape}")

# 8. 출력 이미지 표시
output_image_np = output_image.detach().cpu().numpy()  # 그래디언트 분리 후 numpy 배열로 변환
output_image_np = (output_image_np * 255).astype(np.uint8)  # 0~1 범위 -> 0~255

# 이미지 출력
plt.imshow(output_image_np, cmap='gray')
plt.axis('off')  # 축 제거
plt.show()

# 9. 결과 이미지 저장
output_image_pil = Image.fromarray(output_image_np)  # numpy 배열을 이미지로 변환
output_image_pil.save('output_image.jpg')  # 이미지 파일 저장

# 10. 결과 이미지 다운로드
files.download('output_image.jpg')  # Colab에서 파일 다운로드
```
