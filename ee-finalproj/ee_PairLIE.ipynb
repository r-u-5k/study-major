```python
from google.colab import drive
drive.mount('/content/drive')

import os
import torch
from PIL import Image
import numpy as np
from torchvision import transforms
import matplotlib.pyplot as plt
from google.colab import files
from net.net import net

# 이미지 업로드
uploaded = files.upload()

image_path = list(uploaded.keys())[0]
image = Image.open(image_path)
# image = Image.open('/content/drive/MyDrive/test.jpg')

# 이미지를 RGB로 변환하고 numpy 배열로 변환
image = image.convert('RGB')  # 이미지를 RGB로 변환
image_np = np.array(image)

# 이미지를 tensor로 변환 (전처리)
transform = transforms.Compose([
    transforms.ToTensor(),           # tensor 변환 (값 범위: 0~1)
])
image_tensor = transform(image).unsqueeze(0)  # (1, 3, H, W)

# 모델 정의 및 불러오기
model = net().cpu()
model_path = '/content/drive/MyDrive/PairLIE/weights/epoch_400.pth'
model.load_state_dict(torch.load(model_path))
model.eval()

# 모델에 입력하여 결과 얻기
image_tensor = image_tensor.cpu()
L, R, X = model(image_tensor)
I = torch.pow(L, 0.2) * R
output_image = I

# 출력 이미지 처리
if not os.path.exists("/content/PairLIE/result/"):
    os.mkdir("/content/PairLIE/result/")

I = I.cpu()
I_img = transforms.ToPILImage()(I.squeeze(0))
I_img.save("/content/PairLIE/result/test_result.jpg")
```
